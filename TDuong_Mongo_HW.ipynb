```python
# Exercise 3: Aggregation Pipeline
```

```python
# 1. Write an aggregation pipeline that calculates the average IMDb rating for each genre
# (Return the top 5 genres).
avg_rating_by_genre = collection.aggregate([
 {"$unwind": "$genres"},
 {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
 {"$sort": {"avg_rating": -1}},
 {"$limit": 5}
])

for movie in avg_rating_by_genre :
    pprint.pprint(movie)
```

```python
# 2. Write an aggregation pipeline to find the top 5 directors by the average IMDb rating of
# their movies

top_directors = collection.aggregate([
 {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
 {"$sort": {"avg_rating": -1}},
 {"$limit": 5}
])

for movie in top_directors :
    pprint.pprint(movie)
```

```python
# 3. Write an aggregation pipeline to calculate the total number of movies released in each
# year (Sort the results by the year).

movies_per_year = collection.aggregate([
 {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
 {"$sort": {"_id": 1}}
])

#Limiting 5 to print and check
movies_per_year = collection.aggregate([
 {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
 {"$sort": {"_id": 1}}
])
for movie in movies_per_year  :
    pprint.pprint(movie)
```

```python
# 1. Write a query to update the IMDb rating of a movie with the title "The Godfather" to 9.5

update = collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})
                             
Godfather_updated = collection.find({"title" : "The Godfather"})                  

for movie in Godfather_updated :
    pprint.pprint(movie)

#imdb rating now equals 9.5
```

```python
# 2. Write a query to update all movies where the genre is "Horror" and set their IMDb rating to 6.0 if it is currently null.
collection.update_many({"genres": "Horror", "imdb.rating": ""}, {"$set":{"imdb.rating": 6.0}})

Horrors_null_count = collection.count_documents({"genres" : "Horror", "imdb.rating" : ""})                  
print (f'Updated number of horror movies with empty imdb ratings: {Horrors_null_count}')
```

```python
# 3. Write a query to delete all movies that were released before the year 1950.

collection.delete_many({"year": {"$lt": 1950}})
Movies_before_1950_count = collection.count_documents({"year" : {"$lt":1950}})            
print (f'Updated number of movies released before the year 1950 after delete: {Movies_before_1950_count}')
```
