```python
#1. Reading the CSV File:

#- Load the provided CSV file into a pandas DataFrame.
import pandas as pd
acc_players_df = pd.read_csv('acc_players-2324F.csv', header=1)
#- Display the first 5 rows of the DataFrame to understand the structure of the data.
display(acc_players_df.head(5))
```

```python
#2. Basic Analysis:
#- Calculate the total points scored by all players combined.
total_points = acc_players_df[['PTS']].sum()
print("Total points scored by all players:")
print(total_points)
```

```python
#- Find the player who has played the most minutes (MP).
most_minutes = acc_players_df[['Player', 'MP']].max()
print (most_minutes)
```

```python
#- Identify the top 5 players in terms of total rebounds (TRB).
top_5_rebounds = acc_players_df[['Player','TRB']].sort_values(by='TRB',ascending=False).head(5)
print(top_5_rebounds)
```

```python
#3. Player Filtering:
#- Create a new DataFrame containing only players who played more than 500 minutes.
plus_500_players_df = acc_players_df[(acc_players_df['MP'] > 500)]
display(plus_500_players_df)
```

```python
#- From this filtered DataFrame, determine the player with the highest total assists (AST).
most_assists = plus_500_players_df[['Player', 'AST']].max()
print(most_assists)
```

```python
#- Who are the top 3 Assist Leads in the League?
assist_leads = acc_players_df[['Player','AST']].sort_values(by='AST',ascending=False).head(3)
print(assist_leads)
```

```python
#- Who are the top 3 Shot Blockers?
top_blockers = acc_players_df[['Player','BLK']].sort_values(by='BLK',ascending=False).head(3)
print(top_blockers)
```

```python
#4. School-Based Analysis:
#- Group the players by School and calculate the total points scored by each school.
pts_by_school = acc_players_df.groupby('School')[['School','PTS']].sum('PTS')
print(pts_by_school)
```

```python
#- Group the players by School and calculate the total assists (AST) for each team.
assists_by_team = acc_players_df.groupby('School')[['School','AST']].sum('AST')
print(assists_by_team)
```

```python
#- Sort the schools by total points scored and display the top 3 schools.
top_schools = acc_players_df.groupby('School')[['School', 'PTS']].sum('PTS')
top_schools = top_schools.reset_index()
top_schools = top_schools.sort_values(by='PTS',ascending=False).head(3)
display(top_schools)
```
